```python
from scipy import stats
import pandas as pd

# Assuming you have a DataFrame named df with the specified columns
# Replace 'df' with the actual name of your DataFrame

# Get unique commodity names
commodities = df2

# Create an empty list to store t-test results
t_test_results = []

# Iterate through each commodity
for commodity in commodities:
    # Extract values for the current commodity
    values_2018_20 = df2.loc[(df2['Commodity'] == commodity) & (df2['Date'].between('2018-01-01', '2020-12-31')), 'Arrivals']
    values_2021_23 = df2.loc[(df2['Commodity'] == commodity) & (df2['Date'].between('2021-01-01', '2023-12-31')), 'Arrivals']
    
    values_2018_23_D = df2.loc[(df2['Commodity'] == commodity) & (df2['Date'].between('2018-01-01', '2023-12-31')), 'District']
    values_2018_20.dropna(inplace=True)
    values_2021_23.dropna(inplace=True)
    Districts=values_2018_23_D.unique()
    Avg_arr_before = values_2018_20.sum()/3
    Avg_arr_after = values_2021_23.sum()/3
   
    # Perform t-test
    t_statistic, p_value = stats.ttest_ind(values_2018_20, values_2021_23, nan_policy='omit')

    # Calculate the percentage change
    percentage_change = ((Avg_arr_after - Avg_arr_before) / Avg_arr_before) * 100

    # Determine inference based on p-value
    if p_value < 0.05:  # You can adjust the significance level as needed
        inference = 'Significant'
    else:
        inference = 'Not Significant'

    # Append results to the list
    t_test_results.append({
        'Crop': commodity,
        'Avg_before': Avg_arr_before,
        'Avg_after': Avg_arr_after,
        'Percentage_Change': percentage_change,
        'p_value': p_value,
        'Inference': inference,
        'District':Districts
    })

# Create a DataFrame from the list
t_test_results_df = pd.DataFrame(t_test_results)
```
